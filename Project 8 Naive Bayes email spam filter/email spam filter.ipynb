```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
df = pd.read_csv('email.csv')
df.head()
```

```python
df = df[df.Category != '{"mode":"full"']
df.head()
```

```python
dummydata = pd.get_dummies(df.Category)
dummydata
```

```python
spam = dummydata.spam
df = pd.concat([df,spam],axis = 'columns')
df = df.drop('Category', axis = 1)
df.head()
```

```python
X = df.Message
X_train, X_test, y_train, y_test = train_test_split(X,spam,train_size = 0.8)
```

```python
countVectorizer = CountVectorizer()
```

```python
countVectorizer.fit(X_train.values)
X_train_vector =  countVectorizer.transform(X_train.values)
```

```python
X_train_vector.toarray()[:3]
```

```python
model = MultinomialNB()
```

```python
model.fit(X_train_vector,y_train)
```

```python
model.score(countVectorizer.transform(X_test.values),y_test)
```

```python
## using sklearn pipelined
from sklearn.pipeline import Pipeline
Clf = Pipeline([
    ('vectorizer',CountVectorizer()),
    ('nb',MultinomialNB())
])
```

```python
Clf.fit(X_train,y_train)
```

```python
Clf.score(X_test,y_test)
```
